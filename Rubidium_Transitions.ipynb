```python
import matplotlib.pyplot as plt
```

```python
#RUBIDIUM 85
d1_85 = 377.107385690e12 #Hz pm46
s2_85 = 1.7708439228e9 #Hz pm35
s3_85 = 1.2648885163e9 #Hz pm25
p2_85 = 210.923e6 #Hz pm99 
p3_85 = 150.659e6 #Hz pm11

F22_85 = d1_85 + s2_85 - p2_85  #from S F=2 to P F=2
F23_85 = d1_85 + s2_85 + p3_85  #from S F=2 to P F=3
F32_85 = d1_85 - s3_85 - p2_85
F33_85 = d1_85 - s3_85 + p3_85


#print("85Rb energy levels" , f"{s2p2_85:.15e}, {s2p3_85:.15e}, {s3p2_85:.15e}, {s3p3_85:.15e}")
#transitionlabels_85= ['22', '23', '32', '33']
Rb_85_labels= ['32', '33', '22', '23']
Rb_85 = [F32_85, F33_85, F22_85, F23_85]
print("expected transitions 85Rb", Rb_85)
print(Rb_85_labels)
```

```python
#RUBIDIUM 87
d1_87 = 377.1074635e12 #Hz pm4
s1_87 = 4.27167663181519e9 #Hz pm6 
s2_87 = 2.56300597908911e9 #Hz pm4 
p1_87 = 510.410e6 #Hz pm19 
p2_87 = 306.246e6 #Hz pm11

F11_87 = d1_87 + s1_87 - p1_87  
F12_87 = d1_87 + s1_87 + p2_87
F21_87 = d1_87 - s2_87 - p1_87
F22_87 = d1_87 - s2_87 + p2_87

print("87Rb energy levels", f"{s1p1_87:.15e}, {s1p2_87:.15e}, {s2p1_87:.15e}, {s2p2_87:.15e}")

Rb_87_labels= ['21', '22', '11', '12']
Rb_87 = [ F21_87, F22_87, F11_87, F12_87]
print("expected transitions 87Rb", Rb_87)
```

```python
# Create the plot
plt.figure(figsize=(10, 3))
plt.scatter(Rb_87, [0, 0, 0, 0], label='Transitions 87')  # Scatter plot 
plt.scatter(Rb_85, [0, 0, 0, 0], label='Transitions 85')  # Scatter plot 

# Add labels directly on the graph
for x, label in zip(Rb_87, Rb_87_labels):
    plt.text(x, 0.005, label, ha='center', fontsize=10 ) 
for x, label in zip(Rb_85, Rb_85_labels):
    plt.text(x, 0.005, label, ha='center', fontsize=10 ) 

# Customize the plot
plt.title('Expected Transitions Rb87 and Rb85')
plt.legend()
plt.xlabel('Transition Frequency (Hz)')
plt.yticks([])  # Remove y-axis ticks since all points are on y=0
#plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  # Add a horizontal line at y=0
plt.grid(axis='x', linestyle='--', alpha=0.7)  # Add vertical grid lines
plt.tight_layout()

# Show the plot
plt.show()
```

```python
Rb_85_labels_cross= ['32', 'cross', '33', '22', 'cross', '23']
Rb_87_labels_cross= [ '21', 'cross','22', '11', 'cross', '12']

cross_85_2 = (Rb_85[0] + Rb_85[1] )/2 
cross_85_3 = (Rb_85[2] + Rb_85[3] )/2 
cross_87_1 = (Rb_87[0] + Rb_87[1] )/2 
cross_87_2 = (Rb_87[2] + Rb_87[3] )/2 

Rb_85_cross = [Rb_85[0], cross_85_2, Rb_85[1], Rb_85[2], cross_85_3, Rb_85[3]]
Rb_87_cross = [Rb_87[0], cross_87_1, Rb_87[1], Rb_87[2], cross_87_2, Rb_87[3]]

print(Rb_85_cross)
print(Rb_87_cross)
```

```python
# Create the plot
plt.figure(figsize=(10, 3))
plt.scatter(Rb_87_cross, [0, 0, 0, 0, 0, 0], label='Transitions 87')  # Scatter plot 
plt.scatter(Rb_85_cross, [0, 0, 0, 0, 0, 0], label='Transitions 85')  # Scatter plot 

# Add labels directly on the graph
for x, label in zip(Rb_87_cross, Rb_87_labels_cross):
    plt.text(x, 0.005, label, ha='center', fontsize=8 ) 
for x, label in zip(Rb_85_cross, Rb_85_labels_cross):
    plt.text(x, 0.005, label, ha='center', fontsize=8 ) 

# Customize the plot
plt.title('Expected Transitions Rb87 and Rb85 + crosspeaks')
plt.legend()
plt.xlabel('Transition Frequency (Hz)')
plt.yticks([])  # Remove y-axis ticks since all points are on y=0
#plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  # Add a horizontal line at y=0
plt.grid(axis='x', linestyle='--', alpha=0.7)  # Add vertical grid lines
plt.tight_layout()

# Show the plot
plt.show()  
```
